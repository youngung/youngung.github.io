```python
A = [[1, 2], [3, 4]]
B = [[2, 0], [1, 3]]

C=[[0,0],[0,0]]
C[0][0]=A[0][0]*B[0][0]+A[0][1]*B[1][0]
C[0][1]=A[0][0]*B[0][1]+A[0][1]*B[1][1]
C[1][0]=A[1][0]*B[0][0]+A[1][1]*B[1][0]
C[1][1]=A[1][0]*B[0][1]+A[1][1]*B[1][1]
print('1:',C)

C=[[0,0],[0,0]]
i=0;j=0
C[i][j]=A[i][0]*B[0][j]+A[i][1]*B[1][j]
i=0;j=1
C[i][j]=A[i][0]*B[0][j]+A[i][1]*B[1][j]
i=1;j=0
C[i][j]=A[i][0]*B[0][j]+A[i][1]*B[1][j]
i=1;j=1
C[i][1]=A[i][0]*B[0][j]+A[i][1]*B[j][1]
print('2:',C)


#따라서...
C=[[0,0],[0,0]]
for i in range(2):
    for j in range(2):
        C[i][j]=A[i][0]*B[0][j]+A[i][1]*B[1][j] ## 네 statements가 동일함에 주목!
print('3:',C)

#더 줄이면?
C=[[0,0],[0,0]]
for i in range(2):
    for j in range(2):
        for k in range(2):
            C[i][j]+=A[i][k]*B[k][j]
print('5:',C)

import numpy as np
print('6:',np.array(A) @ np.array(B))          # 행렬 곱
print('7:',np.dot(A, B))   # 동일
```

```python
def get_mag(v):
    import math
    return math.sqrt(v[0]**2+v[1]**2+v[2]**2)

def get_ang(a,b):
    import math
    dotprod=0.
    for i in range(3):
        dotprod+=a[i]*b[i]
    costh=dotprod/(get_mag(a)*get_mag(b))
    print(f'costh:{costh}')
    th=math.acos(costh)
    return th

a=[1,0,0]
b=[0,1,0]
angle=get_ang(a,b)
print('ang in radian:', angle)
## angle to degree?
print('ang in degree:', angle*180/3.141592)
```

Vector 내적

```python
import numpy as np
## Numpy로 구현
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.einsum('i,i->',a,b))
print(a@b)
print(np.dot(a,b))
print(np.inner(a,b))
print(np.sum(a*b))
```
